```python
import pandas as pd
!pip install matplotlib
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación

aqui se muestra el total de ventas por tiendas


```python
total_venta_t1 = tienda['Precio'].sum()
total_venta_t2 = tienda2['Precio'].sum()
total_venta_t3 = tienda3['Precio'].sum()
total_venta_t4 = tienda4['Precio'].sum()

print("total de ventas por ordenado por TIENDAS")
print("========================================")
print(f"el total de ventas Tienda 1 es : {total_venta_t1:,}")
print(f"el total de ventas Tienda 2 es : {total_venta_t2:,}")
print(f"el total de ventas Tienda 3 es : {total_venta_t3:,}")
print(f"el total de ventas Tienda 4 es : {total_venta_t4:,}")
print("")
print("")

#Se crea un diccionario de ventas totales de todas las tiendas
ventas_diccionario = {
    "Tienda 1":total_venta_t1,
    "Tienda 2":total_venta_t2,
    "Tienda 3":total_venta_t3,
    "Tienda 4":total_venta_t4
}

def ordenar_ventas(diccionario_de_ventas):
  #ordeno las ventas de menor a mayor usando funciones lambda
  ventas_ordenadas_por_tienda = sorted(diccionario_de_ventas.items(), key = lambda x: x[1])
  print("total de ventas ordenadas por tienda de MENOR a MAYOR")
  print("=====================================================")

  #muestro las ventas ordenadas
  for tienda, venta in ventas_ordenadas_por_tienda:
    print(f"El total de ventas en la {tienda} es : {venta:,}")

  #devuelvo un diccionario ordenado de ventas a travez de esta funcion
  return dict(ventas_ordenadas_por_tienda)

#establesco algunos espacios para mejorar la visualización
print("")
print("")

#Ejecuta la funcion y recupero el diccionario ordenado de ventas devuelto por la funcion ordenar_ventas
ventas_ordenadas_diccionario = ordenar_ventas(ventas_diccionario)

def graficar_ventas(diccionario_de_ventas):
    #pongo en las variables tiendas y ventas la informacion de estas obtenidas del
    #diccionario pasado en el argumento de la funcion.
    tiendas, ventas = zip(*diccionario_de_ventas.items())

    #obtengo el indice de la tienda que tiene la venta mas baja
    min_venta_index = ventas.index(min(ventas))

    #crea el grafico
    plt.figure(figsize=(10, 6))
    #formateo con rojo la tienda que tiene la venta mas baja y de naranjo las otras
    colores = ['red' if i == min_venta_index else 'orange' for i in range(len(ventas))]
    plt.bar(tiendas, ventas, color=colores)

    #Titulo del grafico
    plt.title('Ventas por Tienda')
    #Nombre del eje X
    plt.xlabel('Tienda')
    #Nombre del eje Y
    plt.ylabel('Total de Ventas')

    #muestra los valores sin decimales.
    for i, v in enumerate(ventas):
        plt.text(i, v + 10000, f"${v:,.0f}", ha='center', va='bottom', fontsize=10)

    #muestra el grafico
    plt.tight_layout()
    plt.show()

#ejecuta la funcion y le paso como argumento las ventas ordenadas de menor a mayor
graficar_ventas(ventas_ordenadas_diccionario)
```

# 2. Ventas por categoría

```python
import numpy as np

venta_por_categoria_t1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
venta_por_categoria_t2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
venta_por_categoria_t3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
venta_por_categoria_t4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()

#convierto las series en un diccionario de diccionarios
datos_ventas = {
    "Tienda 1": venta_por_categoria_t1.to_dict(),
    "Tienda 2": venta_por_categoria_t2.to_dict(),
    "Tienda 3": venta_por_categoria_t3.to_dict(),
    "Tienda 4": venta_por_categoria_t4.to_dict(),
}

def mostrar_ventas_formateadas(ventas_por_categoria, nombre_tienda):

    ventas_categoria_formateadas = {categoria: f"{venta:,.0f}" for categoria, venta in ventas_por_categoria.items()}

    max_len_categoria = max(len(categoria) for categoria in ventas_categoria_formateadas.keys())
    max_len_ventas = max(len(venta) for venta in ventas_categoria_formateadas.values())

    print(f"\nVentas por categoría {nombre_tienda}:")
    for categoria, ventas in ventas_categoria_formateadas.items():
        print(f"Categoria : {categoria:<{max_len_categoria}} las Ventas son: ${ventas:>{max_len_ventas}}")


# Llamamos a la función para cada tienda
mostrar_ventas_formateadas(venta_por_categoria_t1, "Tienda 1")
mostrar_ventas_formateadas(venta_por_categoria_t2, "Tienda 2")
mostrar_ventas_formateadas(venta_por_categoria_t3, "Tienda 3")
mostrar_ventas_formateadas(venta_por_categoria_t4, "Tienda 4")

tiendas = list(datos_ventas.keys())
# Obtener las categorías unicas de todas las tiendas
all_categorias = set()
for tienda_data in datos_ventas.values():
    all_categorias.update(tienda_data.keys())
categorias = sorted(list(all_categorias))

num_tiendas = len(tiendas)
num_categorias = len(categorias)
ancho_barra = 0.8 / num_tiendas
indices_categorias = np.arange(num_categorias)

plt.figure(figsize=(12, 7))

for i, tienda in enumerate(tiendas):
    ventas = [datos_ventas[tienda].get(categoria, 0) for categoria in categorias]
    posiciones_barras = indices_categorias + i * ancho_barra
    plt.bar(posiciones_barras, ventas, width=ancho_barra, label=tienda)

plt.xlabel("Categoría del Producto")
plt.ylabel("Ventas (Suma de Precios)")
plt.title("Ventas por Categoría por Tienda")
plt.xticks(indices_categorias + (num_tiendas - 1) * ancho_barra / 2, categorias, rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas_datos = {
    "Tienda 1":tienda,
    "Tienda 2":tienda2,
    "Tienda 3":tienda3,
    "Tienda 4":tienda4
}

print(type(tienda))
print(type(tienda2))
print(type(tienda3))
print(type(tienda4))

plt.figure(figsize=(12, 8))
for i, (nombre_tienda, df) in enumerate(tiendas_datos.items()):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=df, x='Calificación', order=[1, 2, 3, 4, 5])
    plt.title(f'Distribución de Calificaciones - {nombre_tienda}')
    plt.xlabel('Calificación')
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# Promedio de Calificación por Tienda
promedios_calificacion = {}
for nombre_tienda, df in tiendas_datos.items():
    promedios_calificacion[nombre_tienda] = df['Calificación'].mean()

tiendas = list(promedios_calificacion.keys())
promedios = list(promedios_calificacion.values())

plt.figure(figsize=(8, 5))
plt.hlines(tiendas, [0]*len(tiendas), promedios, color='skyblue', linewidth=4)
plt.plot(promedios, tiendas, "o", color='skyblue', markersize=10)
plt.xlabel('Promedio de Calificación')
plt.title('Promedio de Calificación por Tienda')
plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()


```

# 4. Productos más y menos vendidos

```python

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


# Añadir una columna 'Tienda' para identificar el origen de cada venta
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# ya que todos los csv contienen las mismas columnas se
# Concatenaron todos los DataFrames en uno solo
ventas_consolidadas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

print("Primeras filas de los datos consolidados:\n", ventas_consolidadas.head())
print("\nInformación general del DataFrame consolidado:\n", ventas_consolidadas.info())

ventas_totales_por_producto = ventas_consolidadas.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
print("\nVentas Totales por Producto (Consolidado):\n", ventas_totales_por_producto)

# Visualización de los 10 productos con mayores ventas totales
import matplotlib.pyplot as plt
top_10_productos = ventas_totales_por_producto.nlargest(10)
top_10_productos.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Productos con Mayores Ventas Totales (Todas las Tiendas)')
plt.ylabel('Ventas Totales')
plt.xlabel('Producto')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

ventas_por_producto_por_tienda = ventas_consolidadas.groupby(['Producto', 'Tienda'])['Precio'].sum()
print("\nVentas por Producto, por Tienda:\n", ventas_por_producto_por_tienda)

# Visualización de las ventas de un producto específico en cada tienda
producto_ejemplo = ventas_totales_por_producto.index[0]
ventas_producto_ejemplo = ventas_por_producto_por_tienda.loc[producto_ejemplo]
ventas_producto_ejemplo.plot(kind='bar', figsize=(8, 5))
plt.title(f'Ventas de "{producto_ejemplo}" por Tienda')
plt.ylabel('Ventas')
plt.xlabel('Tienda')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


top_productos_por_tienda = ventas_consolidadas.groupby('Tienda').apply(lambda x: x.groupby('Producto')['Precio'].sum().nlargest(5))
print("\nTop 5 Productos Más Vendidos por Tienda:\n", top_productos_por_tienda)

top_productos_por_tienda = top_productos_por_tienda.unstack(level='Tienda')
top_productos_por_tienda.plot(kind='bar', figsize=(12, 7))
plt.title('Top 5 Productos Más Vendidos por Tienda')
plt.ylabel('Ventas')
plt.xlabel('Producto')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()


```

Aqui se ve que solo en el caso del televisor "TV LED UHD 4K" la tienda 4 se queda un poco atras pero en general esta muy parejo con las demas tiendas. Ademas existen dos categorias en las que la tienda 4 no participa vendiendo. Pero en el caso ABXY la tienda 4 es la unica que vende.

# 5. Envío promedio por tienda

```python
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


costo_promedio_envio_t1 = tienda1['Costo de envío'].mean()
costo_promedio_envio_t2 = tienda2['Costo de envío'].mean()
costo_promedio_envio_t3 = tienda3['Costo de envío'].mean()
costo_promedio_envio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio - Tienda 1: ${costo_promedio_envio_t1:.2f}")
print(f"Costo de envío promedio - Tienda 2: ${costo_promedio_envio_t2:.2f}")
print(f"Costo de envío promedio - Tienda 3: ${costo_promedio_envio_t3:.2f}")
print(f"Costo de envío promedio - Tienda 4: ${costo_promedio_envio_t4:.2f}")


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_promedio = [costo_promedio_envio_t1, costo_promedio_envio_t2, costo_promedio_envio_t3, costo_promedio_envio_t4]

plt.figure(figsize=(8, 6))
plt.bar(tiendas, costos_promedio, color='skyblue')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio ($)')
plt.title('Costo de Envío Promedio por Tienda')
plt.ylim(bottom=0)  # Asegurar que el eje y comience en 0
plt.tight_layout()
plt.show()
```

# Ultima parte de analisis

Aqui se ve que la tienda 4 es la que menos gasta en envio y por lo tanto, se puede deducir que es la tienda que menos vende. Si hay que cerrar una tienda entonces la tienda 4 es la que tiene que cerrar.
